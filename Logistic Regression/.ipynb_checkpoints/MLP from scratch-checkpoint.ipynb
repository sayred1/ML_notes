# Forward propagation:
$$\begin{eqnarray}\\
z_1 &= w_1\cdot x + b_1\\
h_1 & = \sigma(z_1)\\
z_2 &= w_2\cdot h_1 + b_2\\
h_2 & = \sigma(z_2)\\
z_3 &= w_3\cdot h_2 + b_3\\
y\_ &= \sigma{(z_3)} \\
\end{eqnarray}$$

# Activation function:
<h1><center>Sigmoid<center></h1>
$$\begin{eqnarray}\\
\sigma(z) &= \frac{1}{1-e^{-z}}\\
\end{eqnarray}$$
  $$\begin{eqnarray}
\sigma'(z) = \frac{\partial \sigma(z)}{\partial z} &= (\frac{1}{1-e^{-z}})^2 e^{-z} = \frac{e^{-z}}{1-e^{-z}}\frac{1}{1-e^{-z}} = \sigma(z)[1-\sigma(z)]
\end{eqnarray}\\$$

<h1><center>Tanh<center></h1>

$$\begin{eqnarray}\\
\sigma(z) &= \frac{e^z - e^{-z}}{e^z+e^{-z}} = \frac{1-e^{2z}}{1+e^{2z}}=1 - \frac{2}{e^{2z}+1}\\
\end{eqnarray}\\$$
  $$\begin{eqnarray}
\sigma'(z) = \frac{\partial \sigma(z)}{\partial z} &= 4e^{2z} (\frac{1}{1+e^{2z}})^2 = \frac{4e^{2z}}{1+e^{2z}}\frac{1}{1+e^{2z}} = 1- \frac{1-e^{2z}}{1+e^{2z}}\frac{1-e^{2z}}{1+e^{2z}} = 1-\sigma(z)^2
\end{eqnarray}$$

<h1><center>Relu<center></h1>
$$\begin{eqnarray}\\
\sigma(z) &= max(0,z)\
\end{eqnarray}\\$$
$$\begin{eqnarray}
\sigma'(z) &= 0, ~~~~\text{for} ~ \sigma(z) = 0
\end{eqnarray}\\$$
$$\begin{eqnarray}
\sigma'(z) &= 1, ~~~~\text{for} ~ \sigma(z) > 0 
\end{eqnarray}\\$$
  
 <h1><center>Softmax<center></h1>
 ### Here i = example index of all probability vectors $z_{i}$ or logit,  k = class k out of C total classes or logits
 ### To show the derivation, subscripts are necessary 
 $$\begin{eqnarray}\\
 \sigma(z_i)=\frac{e^{z_i}}{\sum_{k\in{C}}e^{z_k}}
 \end{eqnarray}\\$$
$$\begin{eqnarray}\\
\sigma'(z_i) = \frac{e^{z_i}\sum_{k\in C}e^{z_k}-e^{z_i}e^{z_i}}{(\sum_{k\in C}e^{z_k})^2} = \frac{e^{z_i}}{\sum_{k\in C}e^{z_k}}(1-\frac{e^{z_i}}{\sum_{k\in C}e^{z_k}}) = \sigma(z_i)[1-\sigma(z_i)]
 \end{eqnarray}\\$$

  
# Loss function:
### y = target, y_ = prediction, m = number of samples
<h1><center>MSE<center></h1>
$$\begin{eqnarray}\\
L = \frac{1}{2m}(y-y\_)^2
\end{eqnarray}\\$$
  $$\begin{eqnarray}\\
  L' = \frac{\partial L}{\partial y\_}=y\_-y
  \end{eqnarray}\\$$ 


<h1><center>Log_Loss<center></h1>
### To show the derivation, subscripts are necessary: the subscripts i, j and k serve as neurons for the output layer.
### The derivative of the log loss function is applicable both for sigmoid and softmax activation functions. 
$$\begin{eqnarray}\\
L &= \frac{1}{m}\sum_{i\in C} y_i \log {y\__i}\\
\end{eqnarray}\\$$ 


$$\begin{eqnarray}\\
L' = \frac{\partial L}{\partial z_i} = \sum_{i\in C}\frac{\partial L}{\partial y\__j} \frac{\partial y\__j}{\partial z_i} 
\end{eqnarray}\\$$ 
### This must be split up into two cases: i = j and i != j:
### i = j:
$$\begin{eqnarray}\\
\frac{\partial L}{\partial y\__j} = -\frac{y_j}{y\__j} ~~~~~~~~~~~~ \frac {\partial y\__j}{\partial z_j}= \sigma(z_i)[1-\sigma(z_i)] = y\__j(1-y\__j)
\end{eqnarray}\\$$ 
### Thus, 
$$\begin{eqnarray}\\
\frac{\partial L}{\partial y\__j} \frac{\partial y\__j}{\partial z_i}  = y_jy\__j-y_j
\end{eqnarray}\\$$ 
### i != j:
$$\begin{eqnarray}\\
\frac{\partial L}{\partial y\__j} = -\frac{y_i}{y\__i} ~~~~~~~~ \frac{\partial y\__j}{\partial z_i} = -\frac{e^{z_i}e^{z_j}}{(\sum_{k\in C}e^{z_k})^2} = -y\__iy\__j
\end{eqnarray}\\$$ 
  ### Thus
  $$\begin{eqnarray}\\
 \frac{\partial L}{\partial y\__j} \frac{\partial y\__j}{\partial z_i} = y\__jy_i
    \end{eqnarray}\\$$ 
  
  ### Finally, summing everything together,
 $$\begin{eqnarray}\\  
 \frac{\partial L}{\partial z_i} = \sum_{i\in C}\frac{\partial L}{\partial y\__j}\frac{\partial y\__j}{\partial z_i} = \sum_{j\in C}( y_jy\__j-y_j)+\sum_{i\in C}y_iy\__j
 \end{eqnarray}\\$$ 
 ### Here the first term goes to 0? and the sum of y_j = 1
  $$\begin{eqnarray}\\  
  = \sum_{j\in C}(y_iy\__j )-y_j = y\__j \sum_{i\in C}y_i -y_j = y\__j - y_j 
  \end{eqnarray}\\$$ 
# Back propagation  
### Subscripts = layer, superscripts = neuron.

<h1><center>Output Layer<center></h1>
$$\begin{eqnarray}\\
\frac{\partial L}{\partial y\_} & = y\_-y\\\
\end{eqnarray}$$

$$\begin{eqnarray}
\delta^{(3)} &= \frac{\partial L}{\partial z^{(3)}} = \frac{\partial L}{\partial y\_} \frac{\partial y\_} {\partial z^{(3)}} = (y\_-y)\sigma'(z^{(3)}),~~~~ \text{if L = MSE}\\
\end{eqnarray}$$
$$\begin{eqnarray}
\delta^{(3)} &= \frac{\partial L}{\partial y\_} & = y\_-y,~~~~ \text{if L = Log_Loss}\\\ 
\end{eqnarray}$$

$$\begin{eqnarray}
\frac{\partial L}{\partial w^{(3)}}& =h^{(2)}\frac{\partial L}{\partial z^{(3)}} =h^{(2)^T} \cdot \delta^{(3)}  ~~~~~~~ \frac{\partial L}{\partial b^{(3)}}=\sum_{l\in{3}}\frac{\partial L}{\partial z_l^{(3)}} = \sum_{l\in{3}}\delta_l^{(3)}. 
\end{eqnarray}$$
<h1><center> Hidden Layer 3 </center></h1>
$$\begin{eqnarray}\\
\frac{\partial L}{\partial h^{(2)}}=\sum_{l\in3}\frac{\partial L}{\partial z_l^{(3)}}\frac{\partial z_l^{(3)}}{\partial h^{(2)}}=\sum_{l\in3}\frac{\partial L}{\partial y\_} \frac{\partial y\_} {\partial z_l^{(3)}}\frac{\partial z_l^{(3)}}{\partial y\_}= \delta_3 \cdot w^{(3)^T} \\\
\end{eqnarray}$$

$$\begin{eqnarray}
\delta^{(2)} & =\frac{\partial L}{\partial z^{(2)}} = \frac{\partial L}{\partial h^{(2)}}\frac{\partial h^{(2)}}{\partial z^{(2)}} =\delta^{(3)} \cdot w^{(3)^T}\sigma'(z^{(2)})\\\
\end{eqnarray}$$

$$\begin{eqnarray}
\frac{\partial L}{\partial w^{(2)}} & = h^{(1)}\frac{\partial L}{\partial z^{(2)}}= h^{(1)^T} \cdot \delta^{(2)} ~~~~~~~ \frac{\partial L}{\partial b^{(2)}} = \sum_{l\in{4}}\frac{\partial L}{\partial z_l^{(2)}}= \sum_{l\in{4}}\delta_l^{(2)}
\end{eqnarray}$$
<h1><center> Hidden Layer 2 </center></h1>
$$\begin{eqnarray}\\
\frac{\partial L}{\partial h^{(1)}}=\sum_{l\in4}\frac{\partial L}{\partial z_l^{(2)}}\frac{\partial z_l^{(2)}}{\partial h^{(1)}}=\sum_{l\in4}\frac{\partial L}{\partial  h^{(2)}} \frac{\partial  h^{(2)}} {\partial z_l^{(2)}}\frac{\partial z_l^{(2)}}{\partial  h^{(2)}}= \delta_2 \cdot w^{(2)^T} \\\
\end{eqnarray}$$

$$\begin{eqnarray}
\delta^{(1)} & =\frac{\partial L}{\partial z^{(1)}} = \frac{\partial L}{\partial h^{(1)}}\frac{\partial h^{(1)}}{\partial z^{(1)}} =\delta^{(2)} \cdot w^{(2)^T}\sigma'(z^{(1)})\\\
\end{eqnarray}$$

$$\begin{eqnarray}
\frac{\partial L}{\partial w^{(1)}} & = x\frac{\partial L}{\partial z^{(1)}}= x^{T} \cdot \delta^{(1)} ~~~~~~~ \frac{\partial L}{\partial b^{(1)}} = \sum_{l\in{4}}\frac{\partial L}{\partial z_l^{(1)}} = \sum_{l\in{4}}\delta_l^{(1)}
\end{eqnarray}$$

```python
%matplotlib inline

import numpy as np
import sys
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
import sklearn

class my_MLPClassifier(object):
    """
    Basic MultiLayer Perceptron (MLP) neural network.
    Args:
    hidden layer: []
    max_iterations: []
    """
    def __init__(self, hidden, max_iter, 
                 learning_rate, decay_rate, 
                 activation='sigmoid', loss='mse'):
        """
        :param hidden: number of hidden neurons
        :param iterations: how many epochs
        :param learning_rate: initial learning rate
        """
        # initialize input parameters
        self.iterations = max_iter
        self.learning_rate = learning_rate
        self.decay_rate = decay_rate
        self.n_hid1, self.n_hid2 = hidden[0], hidden[1]
        self.activation_method = activation
        self.loss_method = loss
    
    def fit(self, x, y):      
        """
        input: [178, 13] = [example, wine features]
        output: [178, 3] = [example, class probabilities] 
        """
        # Initialize the weights and bias according to the input/target data
        dim_in, dim_out = x.shape[1], 3
        self.w1 = np.random.randn(dim_in, self.n_hid1) # [13,4] -- 52 weights 
        self.w2 = np.random.randn(self.n_hid1, self.n_hid2) # [4,4] -- 16 weights
        self.w3 = np.random.randn(self.n_hid2, dim_out) # [4,3] -- 12 weights
        self.b1 = np.random.randn(1,self.n_hid1) # [1,4] -- 4 biases
        self.b2 = np.random.randn(1,self.n_hid2) # [1,4] -- 4 biases
        self.b3 = np.random.randn(1,dim_out)  # [1,3] -- 3 bias
        loss_hist = [] # track loss for printing
        
        for i in range(self.iterations):
            # forward
            (h1,h2,y_) = self.forward(x)
            cost = self.loss_function(y_,y) 
            loss_hist.append(cost)
            
            # backprop
            
            # output layer, activation = sigmoid
            if self.loss_method == 'mse':
                delta3 = (y_-y) * y*(1-y_)
            elif self.loss_method == 'logloss': 
                delta3 = (y_-y)
            gradw3 = np.dot(np.transpose(h2),delta3)
            gradb3 = np.sum(delta3, axis = 0)
            delta2 = np.dot(delta3, np.transpose(self.w3)) * self.grad_activation(h2)
            gradw2 = np.dot(np.transpose(h1),delta2)
            gradb2 = np.sum(delta2, axis=0)  
            delta1 = np.dot(delta2, np.transpose(self.w2)) * self.grad_activation(h1)
            gradw1 = np.dot(np.transpose(x),delta1)
            gradb1 = np.sum(delta1, axis=0)
            
            # update parameters
            learning_rate = -self.learning_rate * (self.decay_rate**i)
            self.w3 += learning_rate*gradw3
            self.b3 += learning_rate*gradb3
            self.w2 += learning_rate*gradw2
            self.b2 += learning_rate*gradb2
            self.w1 += learning_rate*gradw1
            self.b1 += learning_rate*gradb1
            if i%100==0 or i+1==self.iterations:
                print('activation function: {:s}, loss function: {:s} = {:8.3f}, iterations: {:d}, learning rate: {:6.4f}'
                .format(self.activation_method, self.loss_method, round(cost,3), i, learning_rate))
        self.weights = (self.w3, self.w2, self.w1)
        self.bias = (self.b3, self.b2, self.b1)
        
        # plot the training curve
        plt.title('training curve')
        plt.xlabel('epoch')
        plt.ylabel('loss')
        plt.plot(np.arange(0,self.iterations),loss_hist)
        plt.show()
  
    def loss_function(self,y_,y): 
        m = len(y)
        if self.loss_method == 'logloss':
            y_ /= y_.sum(axis=1)[:, np.newaxis]
            return -1 * np.sum(y * np.log(y_))/len(y)
        elif self.loss_method == 'mse':
            return -0.5 * np.sum(np.power(y-y_,2))/len(y)
        
    def hid_activation(self,x): 
        if self.activation_method == 'sigmoid':
            return 1./(1+np.exp(-x))
        elif self.activation_method == 'tanh':
            return np.tanh(x)
        elif self.activation_method == 'relu':
            return np.maximum(x,0)
        else:
            print('%s not a valid activation function' %(self.activation_method))

    def grad_activation(self,x):
        if self.activation_method == 'sigmoid': 
            return x * (1-x)
        elif self.activation_method == 'tanh':
            return 1 - x**2
        elif self.activation_method == 'relu':
            x[x == 0] = 0
            x[x != 0] = 1
            return x

    def out_activation(self,x): 
        #if self.loss_method == 'logloss':
        return np.exp(x)/np.sum(np.exp(x),axis=1)[:, np.newaxis] 
        """
        elif self.loss_method == 'mse': 
            return 1./(1+np.exp(-x)) 
        """
        
    def forward(self,x):
        z1 = np.dot(x,self.w1)+self.b1
        h1 = self.hid_activation(z1)
        z2 = np.dot(h1,self.w2)+self.b2
        h2 = self.hid_activation(z2)
        z3 = np.dot(h2,self.w3)+self.b3
        y_ = self.out_activation(z3) 
        return(h1,h2,y_)
    
    def predict(self, x):
        (h1, h2, y) = self.forward(x)
        y_ = np.zeros(y.shape[0],int) 
        # array to reshape predictions to (178,)
        for i in range(len(y)):
            for j in range(len(y[i])):
                y[i,j] = round(y[i,j],0)
            y_[i] = np.argmax(y[i])
        return y_
```

```python
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder

# load the data
data=load_wine()
x,Y = data.data, data.target

# scale the data
scaler = StandardScaler()  
scaler.fit(x)  
x0 = scaler.transform(x)

# reshape the target list (178,) to (178,3)
ohe = OneHotEncoder()
y = ohe.fit_transform(np.expand_dims(Y, axis=1)).toarray()
```

```python
max_iter=100
my_mlp = my_MLPClassifier(hidden=[4,4],max_iter=max_iter,learning_rate=0.01,decay_rate=0.99,
                          activation='sigmoid',loss='logloss')
my_mlp.fit(x0,y)
y_ = my_mlp.predict(x0) # predictions in shape (173,3)
count = 0
for i in range(len(y_)):
    if y_[i] == Y[i]:
        count += 1
print('accuracy after {:4d} iterations is: {:6.2f} %'.format(max_iter,100*round(count/len(y_),3)))

# look at distribution of the weights
w1 = my_mlp.weights[0].flatten()
w2 = my_mlp.weights[1].flatten()
w3 = my_mlp.weights[2].flatten()
plt.title('weight distribution')
plt.hist(w1, alpha=0.5, label='w1-'+str(len(w1)))
plt.hist(w2, alpha=0.5, label='w2-'+str(len(w2)))
plt.hist(w3, alpha=0.5, label='w3-'+str(len(w3)))
plt.legend()
plt.show()
```
