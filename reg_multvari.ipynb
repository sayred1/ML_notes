```python
from sklearn.datasets import load_boston
boston = load_boston()
X = boston.data
Y = boston.target
crime = X[:,0]
nox = X[:,4]
rooms = X[:,5]
```

```python
from collections import OrderedDict
import pandas as pd
# for features lets take crime per capita [0], nitric oxide concentration levels [4], avg. num of rooms [5]
boston_data = OrderedDict({'Crime ': crime,
             'NOX ': nox,
             'Avg. Rooms ': rooms, 'Price ': Y*10000})
df = pd.DataFrame(boston_data)
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

fig, (ax1,ax2,ax3) = plt.subplots(1, 3,figsize=(16,4))
ax1.set_ylabel('Housing Cost')
ax1.set_xlabel('Crime per capita')
ax2.set_xlabel('Nitric Oxide Concentration')
ax3.set_xlabel('Num. of Rooms')
ax2.set_title('Housing costs as a function of feature parameters')
ax1.plot(crime,Y*1000,'bo') #row=0, col=0
ax2.plot(nox,Y*1000,'bo') #row=1, col=0
ax3.plot(rooms,Y*1000,'bo') #row=0, col=1
plt.show()
```
